**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

```python
# Customer audience: sales trend over the months (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", color="teal")
plt.title("Customer View: Sales Trend Over the Last Six Months", fontsize=13)
plt.ylabel("Sales ($)")
plt.show()
```

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
# Highlight the lowest sales month
highlight_lowest = df["Sales"].idxmin()

colors_lowest = ["gray" if i != highlight_lowest else "darkorange" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors_lowest)
plt.title("Highlighting the Month with the Lowest Sales", fontsize=13)
plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

**Task 3:** Write one or two lines describing your data story from beginning → middle → end.

Sales kicked off okay in Jan and Feb, with marketing spend also cruising along. -> Things took a bit of a dip in March, and guess what? Marketing spend also went down. -> But then, sales bounced back big from April to June, and it looks like bumping up that marketing spend really helped the sales.

Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/CPE031_FEGALAN/HOA10/energy_consumption.csv')
```

```python
# Group by region and get average
regional_avg_cost = df.groupby('regions')['energy_cost_brl'].mean().reset_index()

highlight_index = regional_avg_cost['energy_cost_brl'].idxmax()
highlight_region = regional_avg_cost.loc[highlight_index]


colors = ["skyblue" if i != highlight_index else "crimson" for i in range(len(regional_avg_cost))]


plt.figure(figsize=(10, 6))
bars = plt.bar(regional_avg_cost['regions'], regional_avg_cost['energy_cost_brl'], color=colors)

#annotation
plt.text(bars[highlight_index].get_x() + bars[highlight_index].get_width()/2., bars[highlight_index].get_height(),
         f'{highlight_region["energy_cost_brl"]:.2f}',
         ha='center', va='bottom', color='black', fontsize=10)


plt.title('Average Energy Cost per Region with Overall Peak Highlighted', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Energy Cost (BRL)', fontsize=12)
plt.ylim(0, regional_avg_cost['energy_cost_brl'].max() * 1.1)


plt.show()
```

This chart reveals the average energy costs across different regions, highlighting that the Southeast region has the highest average energy expenditure compared to others.
