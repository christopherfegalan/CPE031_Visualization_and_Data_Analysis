## Introduction to Data Set

```python
# Example 1
import pandas as pd

mydataset = {
    'animals':["Cat", "Dog", "Rabbit"],
    'height':[13,17,8]
}

myvar = pd.DataFrame(mydataset)

print(myvar)
```

```python
#Example 2
import pandas as pd

a = [1,7,2]
myvar = pd.Series(a)
print(myvar)
print(myvar[0])
```

```python
#Example 3
import pandas as pd

a = [1,7,2]
myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
print(myvar["y"])
```

```python
#Example 4
import pandas as pd
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

```python
#Example 5
import pandas as pd

data = {
    "age": [42, 38, 39],
    "weight": [50,40,45]
}

myvar = pd.DataFrame(data)
print(myvar)

print(myvar.loc[0])
print(myvar.loc[[0,1]])
```

```python
#Example 6
import pandas as pd
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df)
print(df.loc["day2"])
```
